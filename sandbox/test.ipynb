```python
import networkx as nx
import matplotlib.pyplot as plt
graph = nx.Graph()
graph.add_nodes_from(["A", "B", "C", "D", "E", "F"])
graph.add_edges_from([("A", "B"), 
                  ("B", "C"), 
                  ("B", "D"),
                  ("C", "D"),
                  ("A", "E"),
                  ("C", "E"),
                  ("C", "F")]
                )
nx.draw_networkx(graph,node_color='red')
plt.savefig('graph_sample.png')
plt.show()
```
